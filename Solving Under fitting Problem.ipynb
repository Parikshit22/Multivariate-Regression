```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
x_data = pd.read_csv("weightedX.csv")
y_data = pd.read_csv("weightedY.csv")
x_data = x_data.values
y_data = y_data.values
```

```python
plt.scatter(x_data,y_data,cmap='r')
plt.show()
```

```python
reg = LinearRegression()
reg.fit(x_data,y_data)
out = reg.predict(x_data)
plt.scatter(x_data,y_data,color = 'r')
plt.plot(x_data,out)
plt.show()
```

#Now as we can see this line is underfitting the data, so we can add one more dimension to the data to make it more complex so that it can fit well in our model.


```python
x1 = x_data**2
x2 = x_data**3
x3 = x_data**4
x2 = x2.reshape(-1,)
x3 = x3.reshape(-1,)
x1=x1.reshape(-1,)
x = x_data.reshape(-1,)
X = np.stack((x,x1,x2,x3),axis = 1)
reg1 = LinearRegression()
reg1.fit(X,y_data)
pred1 = reg1.predict(X)
plt.scatter(x_data,y_data)
plt.scatter(x_data,pred1)
```
